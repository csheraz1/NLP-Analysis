# TASK 2: PREPROCESS YOUR TEXT

```python
def remove_punctuations(text):
    text =str(text)
    result = "".join([char for char in text if char not in string.punctuation])
    return result

def remove_special_characters(txt):
    txt =str(txt)
    result = re.sub(pattern = "[^\w\s]",
              repl = "",
             string =txt)
    return result

def num_to_word(text):
    letters = str(text).split()
    for i, word in enumerate(letters):
        try:
            letters[i] = num2words(word)
        except:
            letters[i] = word
    processed = ' '.join([str(word) for word in letters])
    return processed

def remove_spaces(txt):
    txt =str(txt)
    result = " ".join(txt.split())
    return result

def to_lower(txt):
    txt =str(txt)
    result = txt.lower()
    return result

def remove_stop_words(txt):
    txt =str(txt)
    stopwords = nltk.corpus.stopwords.words('english')
    word_tokens = nltk.word_tokenize(txt)
    result = [word for word in word_tokens if word not in stopwords]
    processed = ' '.join([str(word) for word in result])
    return processed
def apply_lemmatization(txt):
    txt =str(txt)
    wordnet_lemmatizer = WordNetLemmatizer()
    word_tokens = nltk.word_tokenize(txt)
    lemmatized_word = [wordnet_lemmatizer.lemmatize(word) for word in word_tokens]
    lemmatized_word = ' '.join([str(word) for word in lemmatized_word])
    return lemmatized_word
```

```python
import string

def tast2script(words):
    words = remove_punctuations(words)
    words = remove_special_characters(words)
    words = num_to_word(words)
    words = remove_spaces(words)
    words = to_lower(words)
    words = remove_stop_words(words)
    words = apply_lemmatization(words)
    
    return words 
```

```python
def topics(np, num_topics= 15):
    Task5 = pd.read_csv(np)
    Task5['Review'] = Task5['Review'].apply(tast2script)
    lda_model=LatentDirichletAllocation(n_components=num_topics,learning_method='online',random_state=50,max_iter=2)
    vect = TfidfVectorizer(stop_words=stop_words,max_features=700)
    vect_text=vect.fit_transform(Task5['Review'])
    lda_top=lda_model.fit_transform(vect_text)    
    vocab = vect.get_feature_names_out()
    topics = []
    for i, comp in enumerate(lda_model.components_):
        words = []
        vocab_comp = zip(vocab, comp)
        sorted_words = sorted(vocab_comp, key= lambda x:x[1], reverse=True)[:7]
        for t in sorted_words:
            words.append(t[0])
        topics.append(words)
    res = pd.DataFrame(topics, columns = [f'Word {i+1}' for i in range(7)], index = [f'Topic {i+1}' for i in range(num_topics)])
    return res
```
